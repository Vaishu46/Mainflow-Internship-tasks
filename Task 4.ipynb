# This task involves performing exploratory data analysis on a dataset

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('C:\\Users\\Asus\\Downloads\\USvideos.csv')
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Summary statistics of the dataset
print("\nSummary statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
# Visualization of numerical variables distribution
print("\nDistribution of numerical variables:")
df.hist(bins=20, figsize=(20, 15))
plt.show()
```

```python

# Distribution of categorical variables
print("\nDistribution of categorical variables:")
categorical_cols = ['category_id', 'comments_disabled', 'ratings_disabled', 'video_error_or_removed']
for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python

# Identifying outliers using boxplots
print("\nBoxplot to identify outliers:")
numerical_cols = ['views', 'likes', 'dislikes', 'comment_count']
for col in numerical_cols:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

```python

# Correlation matrix to check relationships between variables
print("\nCorrelation matrix:")
corr_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pairplot to explore relationships between variables
print("\nPairplot of numerical variables:")
sns.pairplot(df[numerical_cols])
plt.show()
```
